### Import libraries

```python
import numpy as np
from PIL import Image, ImageFont, ImageDraw
from PIL.ImageChops import add, subtract, multiply, difference, screen
import PIL.ImageStat as stat

import matplotlib.pyplot as plt 
import matplotlib.image as img 
```

### Reading, saving, and displaying an image using PIL

```python
im = Image.open("parrot.jpg") # read the image, provide the correct path
```

```python
print(im.width, im.height, im.mode, im.format, type(im)) # Provides basic information about the image. 
```

```python
im.save("saved_images/s_parrot.png") # save the image, provide the path to save the image
```

```python
pixel_values = list(im.getdata()) # Collapses the pixel values as a list
print(pixel_values[0:5]) # prints the first 5 pixel values of the image
```

```python
im # display the image
```

### Converting a colored image to a grayscale image using PIL

```python
im_gray = im.convert('L') # convert the RGB color image to a grayscale image
im_gray.save('saved_images/parrot_gray.png') # save the image to disk

pixel_values = list(im_gray.getdata())
print(pixel_values[0:5]) # prints the first 5 pixel values of the image

Image.open("saved_images/parrot_gray.png") # read the grayscale image from disk and show
```

### Separating the RGB channels of an image

```python
im = Image.open("parrot.jpg")
ch_r, ch_g, ch_b = im.split() # split the RGB image into 3 channels: R, G and B

# we shall use matplotlib to display the channels
plt.figure(figsize=(18,6))
plt.subplot(1,3,1); plt.imshow(ch_r, cmap=plt.cm.Reds); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(ch_g, cmap=plt.cm.Greens); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(ch_b, cmap=plt.cm.Blues); plt.axis('off')
plt.tight_layout()
plt.show() # show the R, G, B channels
```

```python
pixel_values = list(ch_r.getdata())
print(pixel_values[0:5]) # prints the first 5 pixel values of the image
```

```python
pixel_values = list(ch_g.getdata())
print(pixel_values[0:5]) # prints the first 5 pixel values of the image
```

```python
pixel_values = list(ch_b.getdata())
print(pixel_values[0:5]) # prints the first 5 pixel values of the image
```

### Combining multiple channels of an image

```python
im = Image.merge('RGB', (ch_g, ch_b, ch_r)) # swap the red, green and blue channels obtained last time with split()
im
```

```python
im = Image.merge('RGB', (ch_b, ch_g, ch_r)) # swap the red and blue channels obtained last time with split()
im
```
